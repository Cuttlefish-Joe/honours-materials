```python
def calculate_iceberg_distances(iceberg_list, iceberg_data):
    """
    Calculate the distance of each iceberg from its initial position at each recorded time.
    
    Parameters:
    iceberg_list: List of iceberg names
    iceberg_data: Dictionary with iceberg data [dates, lats, lons]
    
    Returns:
    Dictionary with iceberg names as keys and distance arrays as values
    """
    from obspy.geodetics import gps2dist_azimuth
    import numpy as np
    
    iceberg_distances = {}
    
    for iceberg in iceberg_list:
        if iceberg in iceberg_data:
            dates = iceberg_data[iceberg][0]
            lats = iceberg_data[iceberg][1]
            lons = iceberg_data[iceberg][2]
            
            if len(lats) > 0 and len(lons) > 0:
                # Initial position (first recorded position)
                initial_lat = lats[0]
                initial_lon = lons[0]
                
                # Calculate distances from initial position
                distances = []
                for i in range(len(lats)):
                    if lats[i] != 0 and lons[i] != 0:  # Skip zero coordinates
                        distance_m, _, _ = gps2dist_azimuth(initial_lat, initial_lon, lats[i], lons[i])
                        distances.append(distance_m / 1000)  # Convert to km
                    else:
                        distances.append(np.nan)  # Use NaN for missing data
                
                iceberg_distances[iceberg] = {
                    'dates': dates,
                    'distances_km': distances,
                    'initial_position': (initial_lat, initial_lon)
                }
                
                print(f"{iceberg}: {len(distances)} distance measurements, max distance: {np.nanmax(distances):.1f} km")
            else:
                print(f"{iceberg}: No valid coordinate data")
                iceberg_distances[iceberg] = None
        else:
            print(f"{iceberg}: Not found in iceberg_data")
            iceberg_distances[iceberg] = None
    
    return iceberg_distances

# Calculate distances for PIG icebergs
iceberg_distances = calculate_iceberg_distances(Icebergs_PIG_all, iceberg_data)
```

```python
def plot_all_iceberg_distances_with_eventes_seasonal_touching(
    iceberg_distances, iceberg_list, PIG_years, PIG_times, station_start_range, station_end_range, iceberg_sizes, savefig=False
):
    """
    Plot all iceberg distances with PIG events histogram binned by season, with bars touching across years,
    and station bin counts centered at each season bin, but only for the range of bins with nonzero PIG events.
    """
    import matplotlib.pyplot as plt
    from datetime import datetime
    import numpy as np
    import pandas as pd

    glacier_date_format = '%Y%j'
    colors = plt.cm.tab10(np.linspace(0, 1, len(iceberg_list)))

    # Convert PIG_times to datetime if needed
    if isinstance(PIG_times[0], str):
        PIG_times_dt = [pd.to_datetime(t) for t in PIG_times]
    else:
        PIG_times_dt = PIG_times

    # Remove timezone info
    PIG_times_dt = [dt.replace(tzinfo=None) for dt in PIG_times_dt]

    # Assign season for each event
    def get_season(dt):
        m = dt.month
        if m in [12, 1, 2]:
            return 'Summer'
        elif m in [3, 4, 5]:
            return 'Autumn'
        elif m in [6, 7, 8]:
            return 'Winter'
        else:
            return 'Spring'

    seasons = ['Summer', 'Autumn', 'Winter', 'Spring']
    season_months = {'Summer': 1, 'Autumn': 4, 'Winter': 7, 'Spring': 10}
    season_colors = {'Summer': 'orange', 'Autumn': 'purple', 'Winter': 'blue', 'Spring': 'green'}

    # Get all years present in the data
    min_year = min(station_start_range + [dt.year for dt in PIG_times_dt])
    max_year = max(station_end_range + [dt.year for dt in PIG_times_dt])
    years = list(range(int(np.floor(min_year)), int(np.ceil(max_year)) + 1))

    # Build a list of all season bins (one per season per year)
    bin_dates = []
    bin_labels = []
    for year in years:
        for season in seasons:
            bin_dates.append(datetime(year, season_months[season], 15))
            bin_labels.append(f"{season} {year}")

    # Count events per bin
    season_counts = []
    for year in years:
        for season in seasons:
            count = sum((dt.year == year) and (get_season(dt) == season) for dt in PIG_times_dt)
            season_counts.append(count)

    # --- Calculate station bin counts for each season bin ---
    station_bin_counts = []
    for dt in bin_dates:
        count = sum((start <= dt.year + dt.month/12 + dt.day/365) and (end >= dt.year + dt.month/12 + dt.day/365)
                    for start, end in zip(station_start_range, station_end_range))
        station_bin_counts.append(count)

    # --- Restrict to range with nonzero PIG events ---
    nonzero_indices = [i for i, c in enumerate(season_counts) if c > 0]
    if nonzero_indices:
        first_idx = min(nonzero_indices)
        last_idx = max(nonzero_indices)
        bin_dates = bin_dates[first_idx:last_idx+1]
        bin_labels = bin_labels[first_idx:last_idx+1]
        season_counts = season_counts[first_idx:last_idx+1]
        station_bin_counts = station_bin_counts[first_idx:last_idx+1]
    else:
        # No PIG events at all, plot nothing
        print("No nonzero PIG seasonal event bins found.")
        return

    # Bar width: about 3 months in days
    bar_width = 90

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

    # Top subplot: iceberg drift (left axis) and seasonal histogram (right axis)
    ax1_dist = ax1
    ax1_pig = ax1_dist.twinx()

    legend_seasons = set()
    for i, (date, count, label) in enumerate(zip(bin_dates, season_counts, bin_labels)):
        season = label.split()[0]
        legend_label = season if season not in legend_seasons else ""
        ax1_pig.bar(date, count, width=bar_width, align='center',
                    color=season_colors[season], label=legend_label, alpha=0.4, edgecolor='black', zorder=1)
        legend_seasons.add(season)

    ax1_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax1_pig.tick_params(axis='y', labelcolor='red')
    handles, labels = ax1_pig.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax1_pig.legend(by_label.values(), by_label.keys(), loc='upper right', fontsize=10)

    # Normalize sizes for plotting (optional: adjust scaling factor as needed)
    sizes = np.array(iceberg_sizes)
    min_size = np.nanmin(sizes)
    max_size = np.nanmax(sizes)
    # Use sqrt scaling for area, and scale to a reasonable marker size range
    marker_sizes = 50 + 128 * (np.sqrt(sizes) - np.sqrt(min_size)) / (np.sqrt(max_size) - np.sqrt(min_size) + 1e-6)

    for i, iceberg in enumerate(iceberg_list):
        if iceberg_distances[iceberg] is not None:
            dates = iceberg_distances[iceberg]['dates']
            distances = iceberg_distances[iceberg]['distances_km']
            datetime_dates = []
            for date in dates:
                try:
                    dt = datetime.strptime(str(date), glacier_date_format)
                    datetime_dates.append(dt)
                except:
                    datetime_dates.append(None)
            valid_indices = [j for j in range(len(distances)) if not np.isnan(distances[j]) and datetime_dates[j] is not None]
            valid_dates = [datetime_dates[j] for j in valid_indices]
            valid_distances = [distances[j] for j in valid_indices]
            if valid_dates:
                # ax1_dist.plot(valid_dates, valid_distances, 'o-', color=colors[i],
                #               linewidth=2, markersize=4, label=f'{iceberg}', zorder=4)
                
                ax1_dist.plot(valid_dates, valid_distances, '-', color=colors[i], linewidth=2, alpha=0.7)
                ax1_dist.scatter(valid_dates, valid_distances, color=colors[i], 
                            s=marker_sizes[i], label=f'{iceberg}', alpha=0.8, edgecolor='k', linewidth=0.5)
                
    ax1_dist.set_ylabel('Distance from initial position (km)')
    ax1_dist.set_ylim(0, 50)
    # ax1_dist.set_title('Iceberg Drift Distances, PIG Events (Seasonal, Touching), and Station Bin Counts')
    ax1_dist.grid(True, alpha=0.3)
    legende = ax1_dist.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Icebergs', 
                    markerscale=2.1, labelspacing=1.6, handletextpad=1.1)
    legende.get_title().set_fontweight('bold')
    ax1_dist.set_zorder(ax1_pig.get_zorder() + 1)
    ax1_dist.patch.set_visible(False)

    # Bottom subplot: same bars and station bin counts
    ax2_stations = ax2
    ax2_pig = ax2_stations.twinx()
        
    legend_seasons2 = set()
    for i, (date, count, label) in enumerate(zip(bin_dates, season_counts, bin_labels)):
        season = label.split()[0]
        legend_label = season if season not in legend_seasons2 else ""
        ax2_pig.bar(date, count, width=bar_width, align='center',
                    color=season_colors[season], label=legend_label, alpha=0.4, edgecolor='black', zorder=1)
        legend_seasons.add(season)    
        
    ax2_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax2_pig.tick_params(axis='y', labelcolor='red')
    handles, labels = ax2_pig.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax2_pig.legend(by_label.values(), by_label.keys(), loc='upper right', fontsize=10)

    # Plot station bin counts centered at each season bin
    ax2_stations.plot(bin_dates, station_bin_counts, 'o-', color='dimgray', linewidth=2, markersize=5, label='Station Bin Counts', zorder=4)
    ax2_stations.set_ylabel('Active Antarctic Stations')
    ax2_stations.set_xlabel('Date')
    ax2_stations.grid(True, alpha=0.3)
    ax2_stations.legend(loc='upper left')

    ax2_stations.set_zorder(ax2_pig.get_zorder() + 1)
    ax2_stations.patch.set_visible(False)

    plt.tight_layout()
    
    if savefig:
        plt.savefig('Iceberg_Distances_PIG_Events_Seasonal_Touching_StationCounts.svg', bbox_inches='tight')
    plt.show()
    
plot_all_iceberg_distances_with_eventes_seasonal_touching(
    iceberg_distances, Icebergs_PIG_all, PIG_years, PIG_times, station_start_range, station_end_range, Icebergs_PIG_all_size, savefig=True)
```

```python
def plot_all_iceberg_distances_with_eventes_monthly(iceberg_distances, iceberg_list, PIG_years, PIG_times, bin_edges):
    """
    Plot all iceberg distances with PIG events histogram binned by month over time, and station bin counts.
    """
    import matplotlib.pyplot as plt
    from datetime import datetime
    import numpy as np

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True,
                                   gridspec_kw={'height_ratios': [3, 1]})

    glacier_date_format = '%Y%j'
    colors = plt.cm.tab10(np.linspace(0, 1, len(iceberg_list)))

    # Convert PIG_years to datetime objects for histogram
    pig_datetime_years = []
    for pig_year in PIG_years:
        year = int(pig_year)
        fraction = pig_year - year
        days_in_year = 366 if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 365
        day_of_year = int(fraction * days_in_year)
        pig_datetime = datetime(year, 1, 1) + pd.Timedelta(days=day_of_year)
        pig_datetime_years.append(pig_datetime)

    # Convert bin_edges (decimal years) to datetime objects for histogram bins
    bin_edges_dt = []
    for be in bin_edges:
        year = int(be)
        fraction = be - year
        days_in_year = 366 if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 365
        day_of_year = int(fraction * days_in_year)
        dt = datetime(year, 1, 1) + pd.Timedelta(days=day_of_year)
        bin_edges_dt.append(dt)

    # Top subplot: iceberg drift (left axis) and PIG events histogram (right axis)
    ax1_dist = ax1
    ax1_pig = ax1_dist.twinx()

    # --- Monthly histogram ---
    # Convert PIG_times to datetime if needed
    if isinstance(PIG_times[0], str):
        PIG_times_dt = [pd.to_datetime(t) for t in PIG_times]
    else:
        PIG_times_dt = PIG_times

    # Make all datetimes offset-naive (remove timezone info)
    PIG_times_dt = [dt.replace(tzinfo=None) for dt in PIG_times_dt]
    bin_edges_dt = [dt.replace(tzinfo=None) for dt in bin_edges_dt]

    # Assign month for each event
    months = range(1, 13)
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    month_colors = plt.cm.get_cmap('tab20', 12).colors

    # Bin by month
    for i, month in enumerate(months):
        month_times = [dt for dt in PIG_times_dt if dt.month == month]
        ax1_pig.hist(month_times, bins=bin_edges_dt, alpha=0.5, color=month_colors[i], label=month_names[i], zorder=1)
    ax1_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax1_pig.tick_params(axis='y', labelcolor='red')
    ax1_pig.legend(loc='upper right', ncol=2, fontsize=10)

    for i, iceberg in enumerate(iceberg_list):
        if iceberg_distances[iceberg] is not None:
            dates = iceberg_distances[iceberg]['dates']
            distances = iceberg_distances[iceberg]['distances_km']
            datetime_dates = []
            for date in dates:
                try:
                    dt = datetime.strptime(str(date), glacier_date_format)
                    datetime_dates.append(dt)
                except:
                    datetime_dates.append(None)
            valid_indices = [j for j in range(len(distances)) if not np.isnan(distances[j]) and datetime_dates[j] is not None]
            valid_dates = [datetime_dates[j] for j in valid_indices]
            valid_distances = [distances[j] for j in valid_indices]
            if valid_dates:
                ax1_dist.plot(valid_dates, valid_distances, 'o-', color=colors[i],
                              linewidth=2, markersize=4, label=f'{iceberg}', zorder=4)
    ax1_dist.set_ylabel('Distance from initial position (km)')
    ax1_dist.set_ylim(0, 200)
    ax1_dist.set_title('Iceberg Drift Distances, PIG Events (Monthly), and Station Bin Counts')
    ax1_dist.grid(True, alpha=0.3)
    ax1_dist.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    ax1_dist.set_zorder(ax1_pig.get_zorder() + 1)
    ax1_dist.patch.set_visible(False)  # Hide the background patch so histogram bars show through

    ax2_stations = ax2
    ax2_pig = ax2_stations.twinx()

    for i, month in enumerate(months):
        month_times = [dt for dt in PIG_times_dt if dt.month == month]
        ax2_pig.hist(month_times, bins=bin_edges_dt, alpha=0.5, color=month_colors[i], label=month_names[i])
    ax2_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax2_pig.tick_params(axis='y', labelcolor='red')
    ax2_pig.legend(loc='upper right', ncol=2, fontsize=10)

    # Calculate bin centers as datetimes
    bin_center_dates = []
    for i in range(len(bin_edges_dt) - 1):
        center_timestamp = bin_edges_dt[i].timestamp() + (bin_edges_dt[i+1].timestamp() - bin_edges_dt[i].timestamp()) / 2
        bin_center_dates.append(datetime.fromtimestamp(center_timestamp))
    ax2_stations.plot(bin_center_dates, bin_counts, 'o-', color='gray', linewidth=2, markersize=5, label='Station Bin Counts', zorder=4)

    ax2_stations.set_ylabel('Active stations within 1000 km')
    ax2_stations.set_xlabel('Date')
    ax2_stations.grid(True, alpha=0.3)
    ax2_stations.legend(loc='upper left')

    ax2_stations.set_zorder(ax2_pig.get_zorder() + 1)
    ax2_stations.patch.set_visible(False)  # Hide the background patch so histogram bars show through

    plt.tight_layout()
    plt.show()

# Example usage:
#plot_all_iceberg_distances_with_eventes_monthly(iceberg_distances, Icebergs_PIG_all, PIG_years, PIG_times, bin_edges)
```

```python
def plot_all_iceberg_distances_with_eventes_monthly_grouped(iceberg_distances, iceberg_list, PIG_years, PIG_times, bin_edges, bin_counts):
    """
    Plot all iceberg distances with PIG events histogram binned by month (side-by-side bars), and station bin counts.
    """
    import matplotlib.pyplot as plt
    from datetime import datetime
    import numpy as np

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True,
                                   gridspec_kw={'height_ratios': [3, 1]})

    glacier_date_format = '%Y%j'
    colors = plt.cm.tab10(np.linspace(0, 1, len(iceberg_list)))

    # Convert bin_edges (decimal years) to datetime objects for histogram bins
    bin_edges_dt = []
    for be in bin_edges:
        year = int(be)
        fraction = be - year
        days_in_year = 366 if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 365
        day_of_year = int(fraction * days_in_year)
        dt = datetime(year, 1, 1) + pd.Timedelta(days=day_of_year)
        bin_edges_dt.append(dt)

    # Convert PIG_times to datetime if needed
    if isinstance(PIG_times[0], str):
        PIG_times_dt = [pd.to_datetime(t) for t in PIG_times]
    else:
        PIG_times_dt = PIG_times

    # Make all datetimes offset-naive (remove timezone info)
    PIG_times_dt = [dt.replace(tzinfo=None) for dt in PIG_times_dt]
    bin_edges_dt = [dt.replace(tzinfo=None) for dt in bin_edges_dt]

    # Assign month for each event
    months = range(1, 13)
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    month_colors = plt.cm.get_cmap('tab20', 12).colors

    # Calculate counts per bin per month
    bin_month_counts = np.zeros((len(bin_edges_dt) - 1, 12), dtype=int)
    for i, (start, end) in enumerate(zip(bin_edges_dt[:-1], bin_edges_dt[1:])):
        for j, month in enumerate(months):
            count = sum((dt >= start) and (dt < end) and (dt.month == month) for dt in PIG_times_dt)
            bin_month_counts[i, j] = count

    # Calculate bin centers for bar positions
    bin_centers = []
    for i in range(len(bin_edges_dt) - 1):
        center_timestamp = bin_edges_dt[i].timestamp() + (bin_edges_dt[i+1].timestamp() - bin_edges_dt[i].timestamp()) / 2
        bin_centers.append(datetime.fromtimestamp(center_timestamp))

    # Bar width and offsets for grouped bars
    total_bar_width = (bin_edges_dt[1] - bin_edges_dt[0]).total_seconds() / (60*60*24*30) * 0.8  # ~80% of bin width in months
    bar_width = total_bar_width / 12

    # Top subplot: iceberg drift (left axis) and grouped monthly histogram (right axis)
    ax1_dist = ax1
    ax1_pig = ax1_dist.twinx()

    # Plot grouped bars for each month
    for m in range(12):
        # Offset each month's bar within the bin
        offsets = [bc + pd.Timedelta(days=(m - 5.5) * bar_width * 30) for bc in bin_centers]
        ax1_pig.bar(offsets, bin_month_counts[:, m], width=bar_width*30, align='center',
                    color=month_colors[m], label=month_names[m], alpha=0.7, zorder=1, edgecolor='black')

    ax1_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax1_pig.tick_params(axis='y', labelcolor='red')
    ax1_pig.legend(loc='upper right', ncol=2, fontsize=10)

    for i, iceberg in enumerate(iceberg_list):
        if iceberg_distances[iceberg] is not None:
            dates = iceberg_distances[iceberg]['dates']
            distances = iceberg_distances[iceberg]['distances_km']
            datetime_dates = []
            for date in dates:
                try:
                    dt = datetime.strptime(str(date), glacier_date_format)
                    datetime_dates.append(dt)
                except:
                    datetime_dates.append(None)
            valid_indices = [j for j in range(len(distances)) if not np.isnan(distances[j]) and datetime_dates[j] is not None]
            valid_dates = [datetime_dates[j] for j in valid_indices]
            valid_distances = [distances[j] for j in valid_indices]
            if valid_dates:
                ax1_dist.plot(valid_dates, valid_distances, 'o-', color=colors[i],
                              linewidth=2, markersize=4, label=f'{iceberg}', zorder=4)
    ax1_dist.set_ylabel('Distance from initial position (km)')
    ax1_dist.set_ylim(0, 200)
    ax1_dist.set_title('Iceberg Drift Distances, PIG Events (Monthly, Grouped), and Station Bin Counts')
    ax1_dist.grid(True, alpha=0.3)
    ax1_dist.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    ax1_dist.set_zorder(ax1_pig.get_zorder() + 1)
    ax1_dist.patch.set_visible(False)

    # Bottom subplot: grouped monthly histogram and station bin counts
    ax2_stations = ax2
    ax2_pig = ax2_stations.twinx()
    for m in range(12):
        offsets = [bc + pd.Timedelta(days=(m - 5.5) * bar_width * 30) for bc in bin_centers]
        ax2_pig.bar(offsets, bin_month_counts[:, m], width=bar_width*30, align='center',
                    color=month_colors[m], label=month_names[m], alpha=0.7, edgecolor='black')
    ax2_pig.set_ylabel('Number of\nPIG Events', color='red')
    ax2_pig.tick_params(axis='y', labelcolor='red')
    ax2_pig.legend(loc='upper right', ncol=2, fontsize=10)

    ax2_stations.plot(bin_centers, bin_counts, 'o-', color='gray', linewidth=2, markersize=5, label='Station Bin Counts', zorder=4)
    ax2_stations.set_ylabel('Active stations within 1000 km')
    ax2_stations.set_xlabel('Date')
    ax2_stations.grid(True, alpha=0.3)
    ax2_stations.legend(loc='upper left')

    ax2_stations.set_zorder(ax2_pig.get_zorder() + 1)
    ax2_stations.patch.set_visible(False)

    plt.tight_layout()
    plt.show()

# Example usage:
#plot_all_iceberg_distances_with_eventes_monthly_grouped(iceberg_distances, Icebergs_PIG_all, PIG_years, PIG_times, bin_edges, bin_counts)
```
